# La régression linéaire bayésienne

## Présentation de l'approche bayésienne

Probabilités, 2 mondes différents : Fréquentielles et Bayésiennes
Ces 2 mondes interprètent les probabilités différemment :
	Dans une approche fréquentielle, on va estimer la probabilité qu’un évènement se reproduise si l’on réalise à nouveau une même expérience, en se basant que sur les données de l’expérience.
	Dans une approche bayésienne, au contraire, on estime un degré de confiance, et donc la chance qu’un évènement se reproduise, en émettant une hypothèse sur la potentielle distribution.
Si l’on prend l’exemple d’une régression linéaire (fréquentielle), elle suppose une distribution gaussienne, alors que lors d’un régression linéaire bayésienne, on va estimer la distribution (au travers d’assomptions et du théorème de Bayes => probabilités conditionnelles).
Il faut donc savoir si l’on possède des connaissances sur les données précédentes et quel type de prédiction on souhaite afin de décider quelle approche choisir (bayésienne ou fréquentielle).

-> Les lois a priori peuvent également être choisies en fonction d'un certain principe, comme la symétrie ou la maximisation de l'entropie compte tenu des contraintes

Nous avons vu la régression linéaire en cours, où lorsque l'on utilise une approche fréquentiste, il faut  utilise la méthode des moindres carrés ordinaires (MCO), à travers la somme résiduelle des carrés (RSS) pour estimer les paramètres :

Maintenant que nous savons ce qu'est l'approche bayésienne, et bien nous pouvons utiliser cette approche afin d'optimiser notre régression linéaire, et donc utiliser un algorithme d'optimisation bayésienne : c'est ça la régression linéaire bayésienne. La croyance a postériori est déterminée grâce à la combinaison de la croyance a priori sur les paramètres, et la fonction de vraisemblance des données.

Tandis qu'avec les moindres carrés on va estimer les paramètres, on va ici estimer une distribution pour chaque paramètre, et, par défaut, on va alors choisir la valeur de ce paramètre qui est la plus évidente, ou la valeur moyenne.

On peut également en garder plusieurs et avoir ainsi un intervalle de crédibilité. Plus on aura de données, et plus notre intervalle se resserrera et convergera vers les valeurs d'une régression linéaire fréquentielle. (voir schéma ci-dessous, les lignes de régression se superposent)

L'approche fréquentiste suppose que nous disposons de suffisamment de mesures à disposition pour faire des estimations significatives.

### Challenge analytique : souvent, nous ne pouvons pas calculer la distribution complète (car ce sera le ratio de deux intégrales correspondants au produit de la vraisemblance et de l'a priori) => MCMC

Du coup, on va s'aider de méthodes de sampling afin d'estimer la distribution a postériori de nos paramètres de régression. Un ensemble de méthodes les plus répandues est les Markov Chain Monte Carlo (MCMC).

![Picture title](image-20230321-173937.png)

Monte Carlo sampling consiste à prendre différents samples aléatoires pour obtenir une réponse.

Une chaîne de Markov est un process où chaque état suivant dépend de l'état actuel, il n'y a pas besoin de connaître tous les états antérieurs. (ici l'attribution de valeurs à des paramètres)

Ces deux concepts mis ensemble, nous avons les chaînes de Markov de Monte Carlo (MCMC). Elles permettent de générer des samples aléatoires pour nos paramètres, et de les tester sur les données. On commence par initialiser nos paramètres, puis on vient en générer des aléatoires par la suite. A chaque nouvelle génération de valeur aléatoire, il y a une évaluation, et si cette évaluation conclu qu'elles ne correspondent pas avec nos données, l'état reste inchangé. Ce processus continue sur plusieurs itérations, et plus il fera d'itération, plus nous convergerons vers la réelle distribution. A la fin du processus, notre algorithme nous retournera toutes les valeurs des paramètres qu'il a accepté (via l'évaluation), nous pourrons ensuite en faire la moyenne pour avoir une estimation de la valeur la plus vraisemblable de nos paramètres. (ou garder la distribution)

Il existe de nombreux algorithmes de type Monte Carlo Markov Chain, et nous ne détailleront pas plus ici mais on peut citer par exemple le No-U-Turn Sampler ou le Metropolis Hastings Sampling.

Ainsi, nos prédictions en utilisant une régression linéaire bayésienne seront des distributions de valeurs possibles.

### Avantages :

-> Adapté aux données en temps réel (les données n'ont pas besoin d'être stockées, utilisable dans les cas d'online supervised machine learning, pas besoin de réentraîner le modèle à chaque nouvelle donnée)

-> Pas besoin d'énormément de données

-> Nous pouvons utiliser nos pensées et connaissances a priori, contrairement à l'approche fréquentiste qui estime que tout est à apprendre des données

-> Possibilité de quantifier l'incertitude que nous avons à propos du modèle (du fait des distributions des paramètres)

### Inconvénients :

-> Chronophage (lent car computationnellement lourd, à la fois pour la partie inférentielle que pour la partie entraînement) => la complexité algorithmique va aussi dépendre du nombre d'étapes MCMC.

-> Dans les cas où nous disposons d'une très grande quantité de données, il vaudra mieux se diriger vers l'approche fréquentiste de la régression linéaire, car la distribution des paramètres de l'approche bayésienne tendra vers l'estimation des paramètres de l'approche linéaire inférée par la méthode des moindres carrés.

-> Pas grandement implémentée dans les librairies de machine learning

### Use cases :

Météo, production d'énergie, recherches médicales (par exemple utilisé dans une étude tentant de déterminer des associations entre la sensibilité à la douleur et les différents facteurs chez les femmes atteintes de Fibromyalgia syndrome. Ils ont obtenu de meilleurs résultats qu'avec l'approche fréquentiste)

```
# Imports
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import BayesianRidge, LinearRegression
from sklearn.model_selection import learning_curve
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv("/work/CAR DETAILS FROM CAR DEKHO.csv")
X = data[["year", "km_driven"]]
y = data["selling_price"]
```

```
data
```

```
# Split train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
```

```
# Creating and training model bayesien 
model = BayesianRidge()
model.fit(X_train,y_train)
```

```
# Model making a prediction on test data
prediction = model.predict(X_test)

# Evaluation of r2 score of the model against the test set
print(f"r2 Score Of Test Set : {r2_score(y_test, prediction)}")
```

```
reglin = LinearRegression()

reglin.fit(X_train, y_train)

pred = reglin.predict(X_test)

print(f"r2 Score Of Test Set : {r2_score(y_test, pred)}")
```

```
test_list = []
train_list = []
samples = []

train_size_abs, train_scores, test_scores = learning_curve(
    model, X, y, train_sizes=[0.35, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 0.95], 
    scoring="r2"
    )
for train_size, cv_train_scores, cv_test_scores in zip(
    train_size_abs, train_scores, test_scores):
    samples.append(train_size)
    train_list.append(cv_train_scores.mean())
    test_list.append(cv_test_scores.mean())
```

```
lin_test_list = []
lin_train_list = []
lin_samples = []

train_size_abs, train_scores, test_scores = learning_curve(
    reglin, X, y, train_sizes=[0.35, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 0.95], 
    scoring="r2"
    )
for train_size, cv_train_scores, cv_test_scores in zip(
    train_size_abs, train_scores, test_scores):
    lin_samples.append(train_size)
    lin_train_list.append(cv_train_scores.mean())
    lin_test_list.append(cv_test_scores.mean())
```

```
plt.plot(samples, test_list)
# train en orange, test en bleu
plt.plot(samples, train_list)
plt.title("Learning curve")
plt.xlabel("Samples sizes")
```

```
plt.plot(lin_samples, lin_test_list)
# train en orange, test en bleu
plt.plot(lin_samples, lin_train_list)
plt.title("Learning curve")
plt.xlabel("Samples sizes")
```
